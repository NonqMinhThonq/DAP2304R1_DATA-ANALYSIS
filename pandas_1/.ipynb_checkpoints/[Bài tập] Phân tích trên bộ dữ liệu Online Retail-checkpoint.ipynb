```python
#c b d liu, cho bit s dng, s ct v kiu d liu ca cc thuc tnh.
import pandas as pd
import numpy as np

df = pd.read_csv(r"D:\DAP2304R1_DATA-ANALYSIS\pandas_1\DATA\OnlineRetail.csv", encoding = 'iso-8859-1')
df
```

```python
print('line number of data is:', df.shape[0])
print('number column of data is:', df.shape[1])
```

```python
df.info()
```

```python
#Trch xut d liu cc ct Description v Quantity lu vo file OnlineRetail.csv

a = df[['Description','Quantity']]
a
a.to_csv('online_retail.csv', index = False)
```

```python
#rch xut d liu 1000 dng u tin lu vo file OnlineRetail.xlsx

b = df.loc[:1000,:]
b
b.to_excel('online_retail.xlsx', sheet_name='Sheet1', index = False)
```

```python
#Trch xut cc bn ghi c s lng t 10 tr ln lu vo file OnlineRetail.h5

c = df.loc[df['Quantity'] > 10]
c
c.to_hdf('online_retail.h5', key = 'data', mode = 'w')
```

```python
#Trch xut d liu cc phn t t dng 1000 n dng 2000, cc ct Quantity, InvoiceDate, UnitPrice lu vo file OnlineRetail.json

d = df.loc[1000:2000,['Quantity','InvoiceDate','UnitPrice']]
d

d.to_json('online_retail.json')
```
