```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
import pandas as pd
import numpy as np

df = pd.read_csv(r"D:\DAP2304R1_DATA-ANALYSIS\pandas_1\DATA\OnlineRetail.csv", encoding = 'iso-8859-1')
df
```

```python
print('line number of data is:', df.shape[0])
print('number column of data is:', df.shape[1])
```

```python
df.info()
```

```python
#Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

a = df[['Description','Quantity']]
a
a.to_csv('online_retail.csv', index = False)
```

```python
#rích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx

b = df.loc[:1000,:]
b
b.to_excel('online_retail.xlsx', sheet_name='Sheet1', index = False)
```

```python
#Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5

c = df.loc[df['Quantity'] > 10]
c
c.to_hdf('online_retail.h5', key = 'data', mode = 'w')
```

```python
#Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json

d = df.loc[1000:2000,['Quantity','InvoiceDate','UnitPrice']]
d

d.to_json('online_retail.json')
```

```python
#Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
e = df.loc[df['InvoiceNo'] == '536365']
e

e.to_html('online_retail.html')
```
