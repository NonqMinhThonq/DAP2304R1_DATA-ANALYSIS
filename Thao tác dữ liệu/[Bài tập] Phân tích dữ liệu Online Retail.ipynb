```python
# Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;


df['month'] = pd.to_datetime(df['InvoiceDate']).dt.month
df.head()
```

```python
df = df.rename(columns = {'month':'Month'})
```

```python
conditions = [(df['Month'] == 1) |  (df['Month'] == 2) | (df['Month'] == 3),
              (df['Month'] == 4) |  (df['Month'] == 5) | (df['Month'] == 6),
              (df['Month'] == 7) |  (df['Month'] == 8) | (df['Month'] == 9),
              (df['Month'] == 10) |  (df['Month'] == 11) | (df['Month'] == 12)]
choices = [1,2,3,4]
df['precious'] = np.select(conditions, choices, default ='')
df
```

```python
df['precious'].value_counts()
```

```python
# Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

df['Amount'] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
#Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.
# Sử dụng phương thức apply() 

def udf(value):
    if value['Country'] == 'United Kingdom' and value['precious'] == '4':
        return ('10%')
    elif value['Country'] == 'France':
        return ('5%')
    else:
        return 0
df['Discount'] = df.apply(udf,axis =1)
df
```

```python
# Sử dụng np.where() để tạo cột mới
df['Discount2'] = np.where((df['Country'] == 'United Kingdom') & (df['precious'] == '4'), 10,np.where(df['Country'] == 'France', 5, 0))
df
```

```python
#Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.’ nhận giá trị bằng: Amount – Discount.

df['Total'] = df['Amount'] - df['Discount2']
df
```
