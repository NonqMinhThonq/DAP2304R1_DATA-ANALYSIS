```python
# 2, Xy dng bng Pivot table, vi mi S ha n tnh trung bnh cng s lng cc mt hng theo tng Quc gia. 

df1 = pd.pivot_table(df, values = 'Quantity', index = 'InvoiceNo', columns = 'Country', aggfunc = 'mean')
df1
```

```python
df.pivot_table(values='Quantity', index='InvoiceNo', columns='Country', aggfunc='mean')
```

```python
# 3, Xy dng bng Pivot table, vi mi Khch hng cho bit s lng mua hng ln nht v nh nht theo Kho.
#pivot_table = pd.pivot_table(df, values='S lng', index='Khch hng', columns='Kho', aggfunc=['max', 'min'])

df.pivot_table(values = 'Quantity', index = 'CustomerID',columns = 'StockCode', aggfunc = {max,min})
```

```python
 df.pivot_table(values='Quantity', index='CustomerID', columns='StockCode', aggfunc=['max', 'min'])
```

```python
# 4,Xy dng bng Pivot table, vi mi M kho tnh tng s lng cc mt hng v trung bnh cng gi.


df.pivot_table(values=['Quantity', 'UnitPrice'], index='StockCode', aggfunc={'Quantity': 'sum', 'UnitPrice': 'mean'})
```

```python
# 5,Xy dng bng Pivot table cho bit tng s lng hng bn c ca mi ngy.

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
display(df.info())

df['day'] = df['InvoiceDate'].dt.day
display(df.head())

pivot_table = pd.pivot_table(df, values='Quantity', index='day', aggfunc='sum')
display(pivot_table)
```
