```python
# 2, Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 

df1 = pd.pivot_table(df, values = 'Quantity', index = 'InvoiceNo', columns = 'Country', aggfunc = 'mean')
df1
```

```python
df.pivot_table(values='Quantity', index='InvoiceNo', columns='Country', aggfunc='mean')
```

```python
# 3, Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
#pivot_table = pd.pivot_table(df, values='Số lượng', index='Khách hàng', columns='Kho', aggfunc=['max', 'min'])

df.pivot_table(values = 'Quantity', index = 'CustomerID',columns = 'StockCode', aggfunc = {max,min})
```

```python
 df.pivot_table(values='Quantity', index='CustomerID', columns='StockCode', aggfunc=['max', 'min'])
```

```python
# 4,Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.


df.pivot_table(values=['Quantity', 'UnitPrice'], index='StockCode', aggfunc={'Quantity': 'sum', 'UnitPrice': 'mean'})
```

```python
# 5,Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
display(df.info())

df['day'] = df['InvoiceDate'].dt.day
display(df.head())

pivot_table = pd.pivot_table(df, values='Quantity', index='day', aggfunc='sum')
display(pivot_table)
```
