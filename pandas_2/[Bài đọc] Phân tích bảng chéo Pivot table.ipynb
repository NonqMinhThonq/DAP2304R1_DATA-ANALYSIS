```python
#c b d liu:

import pandas as pd
df = pd.read_csv(r"D:\DAP2304R1_DATA-ANALYSIS\pandas_2\DATA\FoodPrice_in_Turkey.csv")
df
```

```python
#TH1: To bng y  cc thng tin cn phn tch theo hng, ct

#Tnh trung bnh cng gi ca tng mt hng theo nm

df.pivot_table(values='Price',index='ProductName',columns='Year',aggfunc ='mean')
```

```python
#TH2: To bng vi cc thng tin cn phn tch c t chc theo ct

#Tm gi ln nht theo tng nm

df.pivot_table(values='Price', columns='Year',aggfunc ='max')
```

```python
#TH3: To bng vi cc thng tin cn phn tch c t chc theo dng

#Tm gi ln nht ca tng mt hng

df.pivot_table(values='Price', columns ='ProductName',aggfunc ='max')
```

```python
#TH4: To bng vi cc thng tin tnh ton theo nhiu hm ca cng mt gi tr values 

#Tnh gi ln nht, gi nh nht ca tng a im theo thng

df.pivot_table(values='Price', index='Month', columns='Place', aggfunc = {min, max})
```

```python
#TH5: To bng vi cc thng tin tnh ton theo nhiu gi tr values 

#Tnh tng gi v trung bnh cng s thng bn hng ca tng mt hng theo a im

import numpy as np

df.pivot_table(values=['Price','Month'], index='ProductName', columns='Place', aggfunc ={'Price': np.sum,'Month': np.mean})
```
