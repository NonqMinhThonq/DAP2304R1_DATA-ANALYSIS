```python
print('Mỗi châu lục có bao nhiêu quốc gia nằm trong bảng dữ liệu:')

c = df.groupby('Continent').agg({'Country':'count'})
c
```

```python
print('tổng GDP của mỗi châu lục')

e = df.groupby('Continent').agg({'GDP (millions of US$)':'sum'})
e
```

```python
print('Top 10 quốc gia có GDP cao nhất')

f = df.sort_values('GDP (millions of US$)', ascending = False)
f.head(10)
```
